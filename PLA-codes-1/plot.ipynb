```python
from sklearn.manifold import TSNE


def tsne_plot(start, end, name):
    fdt = FeatureDatasetV1(2, name)
    x = fdt.data.numpy()[start:end]
    y = fdt.labels.numpy()[start:end][:,0]
    tsne = TSNE(n_components=2)
    res = tsne.fit_transform(x)
    plt.xticks([])
    plt.yticks([])
    res1 = res[y==1,:]
    plt.scatter(res1[:,0], res1[:,1], c='tab:blue', alpha=0.3, label='Alice')
    res2 = res[y==0]
    plt.scatter(res2[:,0], res2[:,1], c='tab:orange', alpha=0.3, label='Eve')
    plt.legend(loc='upper right')
```

```python
start, end = 7400, 8400
# using original csi
cdt = CSIDatasetV2(2, 'csi')
x = cdt.data.numpy()[start:end]
y = cdt.labels.numpy()[start:end][:,0]
tsne = TSNE(n_components=2)
res = tsne.fit_transform(x)

plt.xticks([])
plt.yticks([])
res1 = res[y==1,:]
plt.scatter(res1[:,0], res1[:,1], c='tab:blue', alpha=0.3, label='Alice')
res2 = res[y==0]
plt.scatter(res2[:,0], res2[:,1], c='tab:orange', alpha=0.3, label='Eve')
plt.legend(loc='upper right')
plt.savefig('figures/Raw.png', dpi=300, bbox_inches='tight')
```
